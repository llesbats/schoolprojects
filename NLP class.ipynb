```python
#on importe les librairies Soup permettent de naviguer dans des pages web...
from bs4 import BeautifulSoup
import requests
import re

def url_to_string(url):
    res = requests.get(url)
    html = res.text
    soup = BeautifulSoup(html, 'html.parser')
    for script in soup(["script", "style", 'aside']):
        script.extract()
    return " ".join(re.split(r'[\n\t]+', soup.get_text()))

text = url_to_string('https://nyti.ms/2QvbsCt')
#Dans la variable text on stocke le texte de l'article du New York Times. 
#Le pipeline suivant restera le même qu'on peut exécuter avec l'une ou l'autre variable
```

```python
#on enleve la ponctuation
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
tokensWoutPunct = tokenizer.tokenize(text1)

# on affiche le nombre total de token
print(len(tokensWoutPunct))

#on affiche la frequence d'apparition de chaque token
freqWoutPunct = nltk.FreqDist(tokensWoutPunct) 

for w in sorted(freqWoutPunct, key=freqWoutPunct.get, reverse=True):
  print (w, freqWoutPunct[w])

#quel problème cela pose-t-il???>>il faut enlever les stopwords!
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords

clean_tokens = tokensWoutPunct[:] 
i=0
sr = stopwords.words('english')
for token in tokensWoutPunct:
    if token in stopwords.words('english'):
        clean_tokens.remove(token)
freqClean = nltk.FreqDist(clean_tokens) 

for w in sorted(freqClean, key=freqClean.get, reverse=True):
    print (w, freqClean[w])
```

```python
#on repart du même set déjà bien nettoyé pour gagner du temps
clean_tokens_min=[x.lower() for x in clean_tokens]


#on enleve a nouveau les stopwords
cleaner_tokens = clean_tokens_min[:] 
i=0
sr = stopwords.words('english')
for token in clean_tokens_min:
    if token in stopwords.words('english'):
        cleaner_tokens.remove(token)
        i+=1

    #on affiche la frequence d'apparition de chaque token
cleaner_tokens_freq = nltk.FreqDist(cleaner_tokens) 

for w in sorted(cleaner_tokens_freq, key=cleaner_tokens_freq.get, reverse=True):
    print (w, cleaner_tokens_freq[w])
```

```python
#on se rend compte que certains mots partagent les même stems
#on essaie de faire les réconciliations syntaxiques qui vont bien

#on passe par les stems

from nltk.stem import *
from nltk.stem.porter import *

stemmer = PorterStemmer()
stem = [stemmer.stem(rootToken) for rootToken in cleaner_tokens]

stem_freq = nltk.FreqDist(stem) 

for w in sorted(stem_freq, key=stem_freq.get, reverse=True):
    print (w, stem_freq[w])
```

```python
#on se rend compte que certains mots partagent les même stems
#on essaie de faire les réconciliations syntaxiques qui vont bien

#on passe par les stems

from nltk.stem import *
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer("english")
stem = [stemmer.stem(rootToken) for rootToken in cleaner_tokens]

stem_freq = nltk.FreqDist(stem) 

for w in sorted(stem_freq, key=stem_freq.get, reverse=True):
    print (w, stem_freq[w])
```

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()

lem = [wordnet_lemmatizer.lemmatize(rootToken) for rootToken in cleaner_tokens]
lem = nltk.FreqDist(lem) 

for w in sorted(lem, key=lem.get, reverse=True):
    print (w, lem[w])
```
